# Comporta ABC articulada em B — condição de abertura 

## Convenção física usada
- Água a montante atua na face vertical **AB** (diagrama de pressões triangular) com uma impulsão resultante horizontal.
- Água atua **por baixo** da placa horizontal **BC** (diagrama de pressões rectangular), gerando impulsão vertical para cima.
- A comporta é rigida e **AB** mantem sempre o mesmo ângulo com **BC**
- O limiar de abertura ocorre quando $F_{AB}\times b_{AB}>F_{BC}\times b_{BC}=0 \Rightarrow \Pi_{H}\times b_{H}>\Pi_{V}\times b_{v}=0$.

## Método do Diagrama de pressões

É necessário dividir a comporta em duas secções para análise. Uma vertical **AB** e uma horizontal **BC**

### Impulsão, braço e momento vertical
Sabemos que independentemente de $h$ a pressão em $A$ é sempre igual a $0$:

$P_A=0$. 

Calculamos a pressão para o Ponto B por:

$P_B=P_A+\gamma_{w}\times h$

Isto resulta num diagrama de pressões triangular com altura $h$ e base $P_B$. Assim a impulsão resultante é:

$\Pi_H=\frac{1}{2}\times P_B\times h\times w$

Onde w é a largura da comporta. O centro de aplicação da Impulsão é no centro gravitico do diagrama de pressões que no caso de um triangulo corresponde a $\frac{2}{3}h$ quando medido do Ponto $A$. Assim, 

$b_{H}=\frac{1}{3}h$

O momento resultante é portanto (assumindo que a comporta tem 1 m de largura, sendo o cálculo equivalente a considerar por metro de largura de comporta):

$M_H=\Pi_H \times b_{H} \Rightarrow M_H=\frac{1}{2} \times \gamma_{w}\times h \times h \times \frac{1}{3}h = \frac{1}{6}\gamma_{w} h^{3}$

### Impulsão, braço e momento horizontal

Análogamente à metodologi anterior, verifica-se que:

$P_B=P_A+\gamma_{w}\times h$

Sendo uma comporta horizontal o diagrama de pressões é um rectângulo com altura $P_B$ e largura $l_{BC}$. Assim a impulsão resultante é:

$\Pi_V= P_B\times l_{BC} \times w$

O centro de aplicação da Impulsão é no centro gravitico do diagrama de pressões que no caso de um rectangulo corresponde a $\frac{1}{2}l_{BC}$ quando medido do Ponto $B$. Assim, 

$b_{V}=\frac{1}{2}l_{BC}$

O momento resultante é portanto (assumindo que a comporta tem 1 m de largura, sendo o cálculo equivalente a considerar por metro de largura de comporta):

$M_V=\Pi_V \times b_{V} \Rightarrow M_V= \gamma_{w}\times h \times l_{BC} \times \frac{1}{2}l_{BC}=\frac{1}{2}\gamma_{w} h l_{BC}^2$

### Situação de equilibrio

Igualando os momentos para obter o equilibrio obtem-se:

$M_H=M_V \Rightarrow \frac{1}{6}\gamma_{w} h^{3}=\frac{1}{2} \gamma_{w} h l_{BC}^2$

Dividindo a equação por $h$ e por $\gamma_{w}$ assumindo que $h$, $\gamma_{w}$, $l_{BC}$ são maiores que $0$, e simplificando a equação, obtem-se:

$\frac{1}{6} h^{2}=\frac{1}{2}l_{BC}^{2} \Rightarrow h^{2}=3 l_{BC}^2 \Rightarrow h= \sqrt {3} l_{BC} \Rightarrow h= 1.5 \sqrt {3}  \:[m] \Rightarrow h \approx 2.598 \:[m] $ 

## Método Geral

Embora se trate de um exercicio académico apenas porque se tratam de comportas verticais e horizontais seguiremos o método geral para demonstração. Salienta-se que este método só pode ser aplicado caso se tenha um fluido único acima da comporta. Aplicamos assim a formulação:

$\Pi=\gamma_w \times S \times h_0$

e 

$b_{H}=h-d+h_{0}$

com

$d=\frac{I_{GG'}}{S\times h_0} sin \alpha$

Onde $S$ representa a área submersa da superficie em análise, $d$ a distância entre o centro de gravidade da superficie e o ponto de actuação da impulsão, $\alpha$ o ângulo que essa mesma superficie faz com a superficie livre do fluido, $h_0$ a distância medida verticalmente do centro de gravidade da superficie até à superficie livre e $I_{GG'}$ o momento de inércia da superficie no eixo que passa pelo centro de gravidade e ao longo da linha de maior declive.

### Impulsão, braço e momento vertical

Aplicando as equações anteriores e assumindo $w=1\:[m]$:

$\Pi_H=\gamma_w \times h \times w \times \frac{h}{2} = \gamma_w \frac{h^2}{2}$

e 

$b_{H}=h-(d+h_{0})=h-(\frac{h^3w}{12} / (h \times w \times \frac{h}{2} ) \times sin\: {90\degree}+ \frac{h}{2})\Rightarrow b_{H}=\frac{h}{3}$

ficando

$M_H=\Pi_H \times b_{H} \Rightarrow M_H=\gamma_w \frac{h^2}{2} \times \frac{h}{3} = \frac{1}{6}\gamma_{w} h^{3}$


### Impulsão, braço e momento horizontal

Aplicando as equações anteriores e assumindo $w=1\:[m]$:

$\Pi_H=\gamma_w \times l_{BC} \times w \times h = \gamma_w h l_{BC}$

e 

$d=\frac{l_{BC}^3 w}{12} / (h \times w \times \frac{l_{BC}}{2} ) \times sin\: {0\degree}=0$

de notar que $d$ é igual a $0$ porque $ sin\: {0\degree}=0$. Note-se também que $d$ não tem impacto no braço uma vez que mede a distância na horizontal. Assim o braço coincide com o centro geometrico da superficie, ficando:

$M_H=\Pi_H \times b_{H} \Rightarrow M_H=\gamma_w h l_{BC} \times l_{BC}/2 = \frac{1}{2} \gamma_w h l_{BC}^{2}$

### Situação de equilibrio

A situação de equilibrio é igual à formulada anteriormente:

$M_H=M_V \Rightarrow \frac{1}{6}\gamma_{w} h^{3}=\frac{1}{2} \gamma_{w} h l_{BC}^2$

Dividindo a equação por $h$ e por $\gamma_{w}$ assumindo que $h$, $\gamma_{w}$, $l_{BC}$ são maiores que $0$, e simplificando a equação, obtem-se:

$\frac{1}{6} h^{2}=\frac{1}{2}l_{BC}^{2} \Rightarrow h^{2}=3 l_{BC}^2 \Rightarrow h= \sqrt {3} l_{BC} \Rightarrow h= 1.5 \sqrt {3}  \:[m] \Rightarrow h \approx 2.598 \:[m] $ 

```python
%matplotlib inline

import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as colors
import ipywidgets as widgets
from ipywidgets import HBox, VBox, Output
from IPython.display import display, Markdown, clear_output
from ipywidgets.widgets.interaction import show_inline_matplotlib_plots

# ---------------- Sliders ----------------

rho_w_w = widgets.FloatSlider(value=1000.0, min=900.0, max=1100.0, step=1.0, description='ρ água (kg/m³)', continuous_update=False)
d_w     = widgets.FloatSlider(value=1.0, min=0.5, max=2.0, step=0.01, description='d', continuous_update=False)
g_w     = widgets.FloatSlider(value=9.81, min=1.0, max=20.0, step=0.01, description='g (m/s²)', continuous_update=False)

L_w     = widgets.FloatSlider(value=1.5, min=0.1, max=5.0, step=0.01, description='L=BC (m)', continuous_update=False)
b_w     = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='largura b (m)', continuous_update=False)

h_w     = widgets.FloatSlider(value=1.0, min=0.0, max=8.0, step=0.01, description='h (m)', continuous_update=False)

ui = VBox([
    HBox([rho_w_w, d_w, g_w]),
    HBox([L_w, b_w, h_w])
])

out_plot = Output()
out_txt  = Output()
display(VBox([ui, out_plot, out_txt]))

# ---------------- Modelo ----------------
# AB (vertical): F_AB = 1/2 rho g b h^2, atua a h/3 acima de B.
# BC (horizontal, água por baixo): p = rho g h, F_BC = rho g h b L, atua em L/2.
# Momento em B: M_AB = -F_AB*(h/3), M_BC = +F_BC*(L/2)

def compute(rho, g, b, h, L):
    FH = 0.5 * rho * g * b * h**2
    M_AB = FH * (h/3 if h>0 else 0.0)

    FV = rho * g * h * b * L
    M_BC = -FV * (L/2)

    M = M_AB + M_BC
    hcrit = math.sqrt(3)*L
    return FH, FV, M_AB, M_BC, M, hcrit


def draw_all(rho, g, b, h, L):
    # Layout: geometria + 2 diagramas separados (AB e BC)
    fig = plt.figure(figsize=(12, 5.2))
    gs = fig.add_gridspec(1, 3, width_ratios=[1.25, 1, 1])

    ax0 = fig.add_subplot(gs[0, 0])
    axAB = fig.add_subplot(gs[0, 1])
    axBC = fig.add_subplot(gs[0, 2])

    # ---------------- Geometria ----------------
    ax0.set_aspect('equal')
    ax0.axis('off')

    # Gate AB e BC
    ax0.plot([0, 0], [0, max(1.2*h, 1.0)], color='black', lw=6)  # AB
    ax0.plot([0, L], [0, 0], color='black', lw=6)               # BC

    # rotula B
    ax0.add_patch(patches.Circle((-0.05, -0.05), 0.1, color='black', ec='black', lw=2))
    ax0.text(-0.3, -0.3, 'B', fontsize=12)
    ax0.text(-0.3, max(1.2*h, 1.0)+0.02, 'A', fontsize=12)
    ax0.text(L, -0.2, 'C', fontsize=12)

    # superfície livre (à esquerda) a y=h
    ax0.plot([-1.2*L, 0], [h, h], color='tab:blue', lw=2)
    #ax0.text(-1.15*L, h+0.05, 'superfície', color='tab:blue', fontsize=10)

    # água à esquerda de AB
    ax0.add_patch(patches.Rectangle((-1.2*L, 0), 1.2*L, h, color='tab:blue', alpha=0.12, lw=0))
    # água por baixo de BC
    #ax0.add_patch(patches.Rectangle((0, -max(0.7*h, 0.6),), L, max(0.7*h, 0.6), color='tab:blue', alpha=0.12, lw=0))
    ax0.add_patch(patches.Rectangle((-1.2*L, -max(0.7*h, 0.6),), 2.4*L, max(0.7*h, 0.6), color='tab:blue', alpha=0.12, lw=0))
    #ax0.text(0.05*L, -max(0.65*h, 0.55), 'água por baixo', color='tab:blue', fontsize=10)

    # cota h
    ax0.annotate('', xy=(0.25*L, h), xytext=(0.25*L, 0), arrowprops=dict(arrowstyle='<->', lw=2))
    ax0.text(0.28*L, 0.5*h, f'h={h:.2f} m', va='center', fontsize=11)
    ax0.text(0.6, 0.25, f'L={L:.2f} m', transform=ax0.transAxes)

    ax0.set_xlim(-1.25*L, 1.25*L)
    ax0.set_ylim(-max(0.9*h, 0.8), max(1.25*h, 1.2))
    #ax0.set_title('Esquema')

    # ---------------- Diagrama AB (triangular) ----------------
    axAB.grid(True, alpha=0.25)
    axAB.set_title('Pressão em AB (triangular)')

    y = np.linspace(0, max(h, 1e-9), 200)
    p_ab = rho*g*(h - y)
    axAB.plot(p_ab, y, color='C0', lw=2)
    axAB.fill_betweenx(y, 0, p_ab, color='C0', alpha=0.15)
    axAB.set_xlabel('p (Pa)')
    axAB.set_ylabel('y (m) acima de B')
    #axAB.set_xlim(10000, 0)
    axAB.invert_xaxis()
    axAB.set_ylim(0, max(h, 0.5))

    # marcar pmax
    pmax = rho*g*h
    axAB.scatter([pmax], [0], color='C0')
    axAB.text(pmax, 0.03*max(h,0.5), 'pmax=ρgh', fontsize=9)

    # ---------------- Diagrama BC (uniforme) ----------------
    axBC.grid(True, alpha=0.25)
    axBC.set_title('Pressão em BC (uniforme)')

    x = np.linspace(0, L, 100)
    p_bc = np.full_like(x, pmax)
    axBC.plot(x, p_bc, color='C1', lw=2)
    axBC.fill_between(x, 0, p_bc, color='C1', alpha=0.15)
    axBC.set_xlabel('x (m) a partir de B')
    axBC.set_ylabel('p (Pa)')
    axBC.set_xlim(0, max(L, 0.1))
    axBC.set_ylim(0, max(pmax, 1.0)*1.05)
    axBC.invert_yaxis()

    plt.tight_layout()
    plt.show()


def update(change=None):
    rho = float(rho_w_w.value) * float(d_w.value)
    g = float(g_w.value)
    L = float(L_w.value)
    b = float(b_w.value)
    h = float(h_w.value)

    FH, FV, M_AB, M_BC, M, hcrit = compute(rho, g, b, h, L)

    with out_plot:
        clear_output(wait=True)
        draw_all(rho, g, b, h, L)
        show_inline_matplotlib_plots()

    with out_txt:
        clear_output(wait=True)

        status = 'ABRE (M>0)' if M>0 else ('FECHA (M<0)' if M<0 else 'LIMIAR (M=0)')
        md = []
        md.append(f"- rho = {rho:.2f} kg/m³, g = {g:.4g} m/s², b = {b:.3g} m")
        md.append(f"- L = {L:.3g} m, h = {h:.3g} m")
        md.append("\n**Forças (resultantes):**")
        md.append(f"- F_AB = 1/2·rho·g·b·h² = {FH:.6g} N")
        md.append(f"- F_BC = rho·g·h·b·lBC = {FV:.6g} N")
        md.append("\n**Momentos em B:**")
        md.append(f"- M_AB = -F_AB·(h/3) = {M_AB:.6g} N·m")
        md.append(f"- M_BC = +F_BC·(lBC/2) = {M_BC:.6g} N·m")
        md.append(f"- M_total = {M:.6g} N·m → **{status}**")
        md.append("\n**Valor crítico:**")
        md.append(f"- h_crit = sqrt(3)·lBC = {hcrit:.6g} m")
        display(Markdown("  \n".join(md)))

for w in [rho_w_w, d_w, g_w, L_w, b_w, h_w]:
    w.observe(update, names='value')

update()
```
