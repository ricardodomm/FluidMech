# Duas comportas de uma barragem

Pretende-se calcular a impulsão total sobre cada comporta da barragem. 

Sabem-se as dimensões da barragem: 

$w=10.0\:[m]$

$h_1=4.5\:[m]$

$h_2=10.5\:[m]$

$h_3=4.2\:[m]$

$\alpha_{C_{1}}= 60 \degree$

E que o ponto $A$, $B$ e $C$ estão todos à mesma cota. O Ponto $A$ situa-se no topo da comporta $C_2$, o Ponto $C$ é o centro da superfície cilíndrica da comporta e o ponto $B$ é o ponto de articulação da comporta que dista $l_{BC}=1.10\:[m]$ de $C$ para a comporta.

<img src="Ex15.png" alt="descrição" width="600">

## Parte 1 - Comporta de charneira superior

Temos que a impulsão em $C_{1}$ é, pelo método geral:

$\Pi_{C_{1}}= \gamma \cdot A \cdot h_{cg}=\gamma \cdot w \cdot l_{C_1} \cdot h_{cg}$

Precisamos de saber a profundidade do centro geométrico $h_{cg}$ e a área $A$, mais especificamente o comprimento molhado da comporta $l_{C_1}$. Assim, usando trignometria:

$sin (\alpha) = h_1/l_{C_1} $ sendo $\alpha=60 \degree$, $l_{C_1}=4.5/sin(60\degree)=5.196\:[m]$

Sendo $h_{cg}=1/2 \cdot h_1=1/2 \cdot 4.5\:[m]=2.25 \:[m]$, substituindo:

$\Pi_{C_{1}}= 9810\:[kg\:m^{-2}\:s^{-2}] \cdot 10\:[m] \cdot 5.196\:[m] \cdot 2.25 \:[m]$

podemos também calcular o valor de $d$.

$d=(I_{GG'}\cdot \:sin\:\alpha)/(A\cdot h_0)$ ou, $d=(l_C^3 \cdot w \cdot 1/12 \cdot \: sin\:\alpha)/(l_C \cdot w \cdot h_{cg})$ 

substituindo:

$d=((5.196\:[m])^3 \cdot 10\:[m] \cdot 1/12 \cdot \: sin\: 60 \degree)/(5.196\:[m] \cdot 10\:[m] \cdot 2.25\:[m])=0.866\:[m]$

## Parte 2 - Comporta inferior

Para o cálculo de superfícies curvas submersas é necessário fazer a divisão da impulsão ($\Pi$) nas suas componentes principais. Assim, devemos dividir neste caso, por se tratar de uma comporta "cilíndrica", nas direcções cartesianas $x$ e $y$ em $\Pi_v$ e $\Pi_h$, vertical e horizontal respectivamente.

### Cálculo da Impulsão horizontal na comporta $\Pi_h$

No cálculo da Impulsão horizontal devemos considera a projecção da impulsão da superficie num plano vertical. Assumindo linhas isobáricas como sendo paralelas, resulta no cálculo da pressão hidrostática vertical.

Assim, pelo método do diagrama de pressões, 

$\Pi_h=(P_A+P_{A'})/2 \cdot A_{\Pi_h}$

Sendo $A'$ o ponto situado verticalmente por baixo do ponto A mas à cota do fundo da comporta. $A_{\Pi_h}$ corresponde à área projectada no plano vertical. Assim, 

$A_{\Pi_h}=h_3 \cdot w = 4.2 \:[m] \cdot 10 \:[m] = 42 \:[m^2]$, 

$P_A=(h_1+h_2)\gamma=(4.5\:[m]+10.5\:[m])\cdot 9810 \:[kg\:m^{⁻2}\:s^{⁻2}]=147.15 \:[kPa]$, e 

$P_{A'}=(h_1+h_2+h_3)\gamma=(4.5\:[m]+10.5\:[m]+4.2\:[m])\cdot 9810 \:[kg\:m^{⁻2}\:s^{⁻2}]=188.352\:[kPa]$, assim:

$\Pi_h=(147.15 \:[kPa]+188.352\:[kPa])/2 \cdot 42 \:[m^2]=7045.54\:[kN]$

### Cálculo da Impulsão vertical na comporta $\Pi_v$

A Impulsão vertical obtém-se calculando o peso de um liquido fictício por cima da comporta até à superfície. 

Assim, precisamos de caracterizar o volume de liquido acima da comporta. Este é composto por 2 secções: 

#### Área 1: $A_1$
A área $A_1$ corresponde ao paralelipipedo entre a superfície livre e uma linha horizontal no ponto $A$. 

Para calcular esta secção temos que primeiro calcular $l_{\over{AD}}$ com $D'$ o ponto onde a comporta encontra o leito e $D$ um ponto na intersecção de uma projecção vertical com a linha $\overline{A\:B\:C}$

<img src="Ex15_Det.png" alt="descrição" width="200">

Sabemos que $l_{\over{AC}}= 6\:[m]$ e que $l_{\over{CD'}}= 6\:[m]$, sabemos também que $l_{\over{DD'}}=h_3=4.2 \:[m]$, pelo teorema de Pitágoras:

$l_{\over{CD'}}^2=l_{\over{DD'}}^2+l_{\over{CD}}^2 \Rightarrow 6^2=4.2^2+l_{\over{CD}}^2 \Rightarrow l_{\over{CD}}= \sqrt{6^2-4.2^2}=4.285\:[m]$ e por conseguinte:

$l_{\over{AD}}=l_{\over{AC}}-l_{\over{CD}} \Rightarrow l_{\over{AD}}=6\:[m]-4.285\:[m] \Rightarrow l_{\over{AD}}=1.715\:[m] $

Assim:

$A_1=l_{\over{AD}} \cdot w = 1.715\:[m] \cdot 15\:[m]= 25.73 \:[m^2]$

Obtendo para o Peso:

$P_1= A_1 \cdot w \cdot \gamma= 25.75\:[m^2]\cdot 10\:[m] \cdot 9810 \:[kg\:m^{-2}\:s^{-2}]= 2523.6 [kN]$

#### Área 2: $A_2$

A área $A_2$ corresponde ao volume delimitado pela superficie inferior de $A_1$ e a comporta. Como se trata de um cilíndro, Podemos obter a área subtraindo à area formada pelo arco $ACD'$, que não é mais que uma porção de uma circunferência, o triangulo $DCD'$.

O primeiro passo é calcular o ângulo ($\theta$) formado por $ACD'$ em $C$. 

Temos $sin\: \theta= h_3/l_{\over{AC}}= 4.2\:[m] / 6\:[m] \Rightarrow \theta=arcsin(4.2/6)=44.43 \degree$

Sendo a área de um circulo $\pi \cdot r^2$ a área do arco $ACD'$ será:

$A_{ACD'}=\pi \cdot r^2 \cdot \theta / 360 \degree=3.1415 \cdot 6^2 \cdot 44.43 \degree / 360 \degree = 13.96 \:[m^2]$

Par ao triângulo $DCD'$ a área será:

$A_{DCD'}=1/2 \cdot l_{\over{CD}}\cdot l_{\over{DD'}} = 1/2 \cdot 4.285\:[m] \cdot 4.2\:[m] = 9.0 \:[m^2]$

e 

$A_2=A_{ACD'}-A_{DCD'}=13.96 \:[m^2]-9.0 \:[m^2] = 4.96 \:[m^2]$

Sendo o peso

$P_2= A_2 \cdot w \cdot \gamma= 4.96\:[m^2]\cdot 10\:[m] \cdot 9810 \:[kg\:m^{-2}\:s^{-2}]= 486.6 [kN]$


#### Peso total

O peso total (Impulsão vertical) é a soma dos dois pesos:

$\Pi_v=P_1+P_2=2523.6\:[kN]+486.6\:[kN]=3010.2\:[kN]$

### Impulsão total

Do teorema de Pitágoras:

$\Pi^2=\Pi_v^2+\Pi_h^2 \Rightarrow \Pi=sqrt{3010.2^2+7045.54^2}=7661.71\:[kN]$

### Somatório de momentos

Para obter a força necessária para abrir a comporta necessitamos de calcular o braço da resultante. Sabemos que a linha de acção da força $\Pi$ passa pelo ponto $C$ por se tratar de uma superficie esférica. Queremos sabe a ditância perpendicular entre a linha de acção de $\Pi$ e a linha de rotação que passa pelo ponto $B$ da comporta.

<img src="Ex15_Det2.png" alt="descrição" width="200">

Sabendo os componentes, e que os componentes da força têm um ângulo igual ao que a linha de acção faz com a horizontal, $\beta=arctan(\Pi_V/\Pi_H)=arctan(3010.2\:[kN]/7045.54\:[kN])=23.2 \degree$

Podemos assim saber a distancia da projecção perpendicular à linha de acção da distância $l_{\over{BC}}$ fazendo:

$sin\:\beta= d_{\Pi}/l_{\over{BC}} \Rightarrow d_{\Pi}=sin\:\beta \cdot l_{\over{BC}}= sin\:23.2 \degree \cdot 1.1\:[m]= 0.432\:[m]$

Do somatório dos momentos em $B$, resulta que:

$F \cdot b_F =\Pi \cdot d_{\Pi} \Rightarrow F =\Pi \cdot d_{\Pi} /(l_{\over{AC}}-l_{\over{BC}}) = 7661.71\:[kN] \cdot 0.432\:[m]/ (6\:[m]-1.1\:[m])=675.48\:[kN]$
